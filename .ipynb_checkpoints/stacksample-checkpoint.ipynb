```python
# Load in the answers
answers_df = pd.read_csv('../Datasets/stacksample/Answers.csv', encoding = 'ISO-8859-1')
answers_df.head(10)
```

## Data Wrangling

#### Let's start with cleaning up the questions and answers

```python
# Create a function to clean up the text
def strip_html_tags(body):
    # Remove html tags
    regex = re.compile('<.*?>')
    no_html_tags = re.sub(regex, '', body)                     
    return no_html_tags
```

```python
# Strip the html tags from the questions
questions_df['Body'] = questions_df['Body'].apply(strip_html_tags)
```

```python
# Have a look at one of the questions
print(questions_df['Body'][1])
```

```python
# Strip the html tags from the answers
answers_df['Body'] = answers_df['Body'].apply(strip_html_tags)
```

```python
# Have a look at one of the answers
print(answers_df['Body'][1])
```
